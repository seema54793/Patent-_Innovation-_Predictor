```python
import os
import requests
import json
from pprint import pprint
from dotenv import load_dotenv
load_dotenv()


api_key = os.getenv("SERP_API_KEY")
if not api_key:
    raise ValueError("API Key is not in environment variables")
```

```python
load_dotenv()  # This looks in the current folder
print("API Key from env:", os.getenv("SERP_API_KEY"))

```

```python
query = "langgraph"
url = f"https://serpapi.com/search?engine=google_patents&q={query}&api_key={api_key}"
```

```python
response = requests.get(url)
if response.status_code == 200:
    data = response.json()
    print(data)
    patents = data.get("organic_results",[])
    with open("files/patents.json","w") as file:
        json.dump(patents,file,indent = 4)
```

```python
patent_url = "https://serpapi.com/search.json?engine=google_patents_details&patent_id=patent%2FCN119226078A%2Fen&api_key=" + api_key
patents_response = requests.get(patent_url)
if patents_response.status_code ==200:
    patent_data = patents_response.json()
    
    with open('files/patent_details.json',"w") as file:
        json.dump(patent_data,file,indent = 4)
```
